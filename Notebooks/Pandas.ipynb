## **Series**

```python
import pandas as pd
```

```python
serie = pd.Series([4, 7, -5, 3])
serie
```

```python
serie.values
```

```python
serie.index
```

```python
serie2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
print(serie2)
print(serie2.index)
```

```python
print(serie2['a'])
```

```python
serie2['d'] = 6
print(serie2)
```

```python
serie2[['c', 'a', 'd']]
```

```python
serie2[serie2 > 0]
```

```python
'b' in serie2
```

## **DataFrame**

```python
dados = {'cidade': ['Fortaleza', 'Fortaleza', 'Fortaleza', 'Quixada', 'Quixada', 'Quixada'],
         'ano': [2000, 2001, 2002, 2000, 2001, 2002],
         'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

df = pd.DataFrame(dados)
df
```

```python
df.head()
```

```python
pd.DataFrame(df, columns=['ano', 'cidade', 'pop'])
```

```python
df
```

```python
df2 = pd.DataFrame(dados, columns=['ano', 'cidade', 'pop', 'idh'], index=['um', 'dois', 'tres', 'quarto', 'cinco', 'seis'])
df2
```

```python
df2['cidade']
```

```python
df2['idh'] = 0.9
df2
```
